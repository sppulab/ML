```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Admission_Predict.csv')
```

```python
data.shape
```

```python
data
```

```python
for i in range(0,500):
    if(data['Chance of Admit '][i]>=0.8):
        data['Chance of Admit '][i] = 1
    else:
        data['Chance of Admit '][i] = 0
```

```python
data.head()
```

```python
X = data.iloc[:,1:-1]
```

```python
X
```

```python
Y = data.iloc[:,-1:]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =  train_test_split(X,Y ,train_size=0.75, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisionTree = DecisionTreeClassifier(criterion='entropy')
```

```python
decisionTree = decisionTree.fit(X_train,Y_train)
```

```python
Y_predict = decisionTree.predict(X_test)
```

```python
Y_predict
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score
```

```python
Accuracy = accuracy_score(Y_test,Y_predict)
Precision = precision_score(Y_test,Y_predict)
Recall = recall_score(Y_test,Y_predict)
```

```python
print("Accuracy : ",Accuracy)
print("Precision : ",Precision)
print("Recall : ",Recall)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report
```

```python
cm = confusion_matrix(Y_test,Y_predict)
print("Confusion Matrix:\n",cm)
```

```python
cm = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [False, True])
cm.plot()
plt.show()
```

```python
print(classification_report(Y_test,Y_predict))
```

```python
from sklearn.tree import plot_tree
```

```python
fig = plt.figure(figsize=(10,5))
plot_tree(decisionTree,fontsize=8,rounded=True,proportion=True,feature_names=X.columns,class_names=True,max_depth=2,filled=True)
```
